```python
def localizacao(df, coluna):
    lista_lat = []
    lista_long = []

    for i in range(len(df)):
        geolocator=Nominatim(user_agent='geoapi')
        endereco = df[coluna].loc[i]
        location=geolocator.geocode(endereco, timeout=3)
        if location:
            lat=location.latitude
            lon=location.longitude
        else:
            lat= None
            lon = None
        lista_lat.append(lat)
        lista_long.append(lon)
        print(f'Index: {i}, Latitude: {lat}, longitude: {lon}')
    return lista_lat, lista_long

```

```python
df['LATITUDE'], df['LONGITUDE'] = localizacao(df, 'Endereco')
```

```python
df = df.dropna(subset=['LATITUDE']).reset_index(drop=True)
df = df[df['LATITUDE'] <= 4]
```

```python
df['BAIRRO_LOC'] = df['BAIRRO_LOC'].str.replace('Pres.', 'Presidente')
df['BAIRRO_LOC'] = df['BAIRRO_LOC'].str.replace('Pref.', 'Prefeito')
```

```python
df['LAT_BAIRRO'], df['LONG_BAIRRO'] = localizacao(df, 'BAIRRO_LOC')
```
